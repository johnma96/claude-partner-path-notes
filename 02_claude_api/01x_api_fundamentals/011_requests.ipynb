```python
# Create an API client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in 1 sentence."
        }
    ]
)

message.content[0].text
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):

    response = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return response.content[0].text



```

```python
messages = []

add_user_message(messages, "What is quantum computing? Answer in 1 sentence.")

answer = chat(messages)

add_assistant_message(messages, answer)

add_user_message(messages, "Write another sentence that expands on the previous answer.")

answer = chat(messages)

add_assistant_message(messages, answer)

messages
```

```python


messages = []

while True:
    user_input = input("User: ")
    add_user_message(messages, user_input)
    
    assistant_response = chat(messages)
    print(f"Assistant: {assistant_response}")
    
    add_assistant_message(messages, assistant_response)
```
