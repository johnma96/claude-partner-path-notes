```python
# Create an API client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in 1 sentence."
        }
    ]
)

message.content[0].text
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system=None, temperature=1.0):

    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }

    if system:
        params["system"] = system
    
    response = client.messages.create(
        **params
    )
    
    return response.content[0].text
```

```python
messages = []

system = """
        Your are a patient math tutor. 
        Do not directly answer the question, 
        but instead ask guiding questions to help the 
        student arrive at the answer on their own. 
        If the student is stuck, provide a hint. 
        Always be encouraging and positive.
        """

add_user_message(messages, "How do I solve 5x+3=2 for x?")

answer = chat(messages, system=system)


answer
```
